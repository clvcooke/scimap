```python
import matplotlib.pyplot as plt
import pandas as pd

# 6. Distribution of Termination Dates
termination_date_counts = joined_df['termination_date'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
termination_date_counts.plot(kind='line', marker='o', color='purple')
plt.title('Distribution of Termination Dates')
plt.xlabel('Termination Date')
plt.ylabel('Number of Terminations')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from collections import defaultdict
from copy import deepcopy
grants_by_fips = defaultdict(list)
for grant in joined_df.to_dict(orient='records'):

    g = deepcopy(grant)
    fips = grant['FIPS']
    del g['FIPS']
    coords = list(g['centroid'].coords)[0]
    del g['centroid']
    g['centroid'] = [coords[0], coords[1]]
    grants_by_fips[grant['FIPS']].append(g)
```

```python
with open("grants_by_fips.json", "w") as f:
    json.dump(grants_by_fips, f)
```
